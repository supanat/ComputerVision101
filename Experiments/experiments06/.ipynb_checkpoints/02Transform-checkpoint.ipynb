```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib notebook 

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import math
```

```python
image = cv2.imread("../../images/trex.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure("Original")
plt.title("Original")
plt.imshow(image)
```

```python
import imutils
```

### Translation

```python
shifted = imutils.translate(image, 0, 100)
plt.figure("Shifted")
plt.imshow(shifted)
```

```python
cv2.imwrite("shifted_trex.png",shifted)
```

### Rotate

```python
rotated = imutils.rotate(image, 90)
plt.figure("rotated")
plt.imshow(rotated)
```

### Resize

```python
resized = imutils.resize(image, width = 100)
plt.figure("resized")
plt.imshow(resized)
```

### Flipping

```python
h_flipped = cv2.flip(image, 1)
v_flipped = cv2.flip(image, 0)

plt.figure('Image Rotation')
plt.subplot(121) #1ROW,2COL,INDEX 1
plt.imshow(h_flipped) 
plt.title('Flip Horizontal')
plt.axis('off')

plt.subplot(122) #1ROW,2COL,INDEX 2
plt.imshow(v_flipped)
plt.title('Flip Vertical')
plt.axis('off')


plt.show()
```
