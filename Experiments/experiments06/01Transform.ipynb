```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline 

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import math
```

```python
image = cv2.imread("../../images/trex.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.title("Original")
plt.imshow(image)
```

### Transform
 https://docs.opencv.org/2.4/modules/imgproc/doc/geometric_transformations.html?#warpaffine

### Translation

```python
dx = 25
dy = 50
M = np.float32([[1, 0, dx], [0, 1, dy], [0, 0, 1]])
M
```

```python
rows,cols,channels = image.shape

M = np.float32([[1, 0, 25], [0, 1, 50]])

shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
plt.imshow(shifted)
```

```python
M = np.float32([[1, 0, -50], [0, 1, -90]])

shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
plt.imshow(shifted)
```

### Rotate

```python
angle = math.radians(5)
M = np.float32([[math.cos(angle), -math.sin(angle), 0],
                [math.sin(angle), math.cos(angle), 0]])
print(M)

rotated = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
plt.imshow(rotated)
```

```python
centerX, centerY = (image.shape[1] // 2, image.shape[0] // 2)
center = centerX ,centerY 
center
```

```python
angle = 45
rot_mat = cv2.getRotationMatrix2D(center,angle,1.0)
rot_mat
```

```python
rotated = cv2.warpAffine(image, rot_mat, (image.shape[1], image.shape[0]),flags=cv2.INTER_LINEAR)
plt.figure('Image Rotation')
plt.imshow(rotated)
plt.show()
```

### Scaling

```python
sx = 0.5
sy = 0.5

M = np.float32([[sx, 0, 0],[0, sy, 0]])
print(M)

scaled = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
plt.imshow(scaled)
```

```python
cv2.imwrite("scaled_trex.png",scaled)
```
