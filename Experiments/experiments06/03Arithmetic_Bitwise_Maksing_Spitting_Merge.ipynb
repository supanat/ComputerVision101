### Arithmetic

```python
M = np.ones(image.shape, dtype = "uint8") * 100
added = cv2.add(image, M)
plt.figure('added')
plt.imshow(added)
```

```python
M = np.ones(image.shape, dtype = "uint8") * 50
subtracted = cv2.subtract(image, M)
plt.figure('subtracted')
plt.imshow(subtracted)
```

### Bitwise

```python
rectangle = np.zeros((300, 300), dtype = "uint8")
cv2.rectangle(rectangle, (25, 25), (275, 275), [255,255,255], -1)
plt.figure('Rectangle')
plt.imshow(rectangle,cmap='gray')
```

```python
circle = np.zeros((300, 300), dtype = "uint8")
cv2.circle(circle, (150, 150), 50, [255,255,255], -1)
plt.figure('Circle')
plt.imshow(circle,cmap='gray')
```

AND: A bitwise AND is true if and only if both pixels are greater than zero.

```python
bitwiseAnd = cv2.bitwise_and(rectangle, circle)
plt.figure('AND')
plt.imshow(bitwiseAnd,cmap='gray')
```

OR: A bitwise OR is true if either of the two pixels are greater than zero.

```python
bitwiseOr = cv2.bitwise_or(rectangle, circle)
plt.figure('OR_')
plt.imshow(bitwiseOr,cmap='gray')
```

XOR: A bitwise XOR is true if and only if either of the two pixels are greater than zero, but not both.

```python
bitwiseXor = cv2.bitwise_xor(rectangle, circle)
plt.figure('XOR_')
plt.imshow(bitwiseXor,cmap='gray')
```

NOT:AbitwiseNOTinvertsthe“on”and“off”pixels in an image.

```python
bitwiseNot = cv2.bitwise_not(circle)
plt.figure('NOT_')
plt.imshow(bitwiseNot,cmap='gray')
```

### Split and Merge

```python
image = cv2.imread("../../images/messi5.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
(R, G, B) = cv2.split(image)
```

```python
plt.figure('G')
plt.imshow(G,cmap='gray')
```

```python
plt.figure('Channels')
plt.subplot(311) 
plt.imshow(R,cmap='gray') 
plt.title('R')
plt.axis('off')

plt.subplot(312) 
plt.imshow(G,cmap='gray')
plt.title('G')
plt.axis('off')

plt.subplot(313) 
plt.imshow(B,cmap='gray')
plt.title('B')
plt.axis('off')
```

```python
merged = cv2.merge([B, G, R])
merged = cv2.cvtColor(merged, cv2.COLOR_BGR2RGB)


print(merged.shape)
plt.figure('merge')
plt.imshow(merged)
```
