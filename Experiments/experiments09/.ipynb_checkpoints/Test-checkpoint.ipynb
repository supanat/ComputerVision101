```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
image = cv2.imread("../../images/coins.png")
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image,cmap='gray')
```

```python
t = 150


def threshold(src_image,threshold):
    width = src_image.shape[1]
    height = src_image.shape[0]
    for i in range(height):
        for j in range(width):
        
            if (src_image[i,j] > t):
                src_image[i,j] = 255
            else:
                src_image[i,j] = 0
    
    return src_image

thresh = threshold(image.copy(),t)
            
plt.imshow(thresh,cmap='gray')
```

```python
blurred = cv2.GaussianBlur(image, (5, 5), 5)
plt.imshow(blurred,cmap='gray')
```

```python
thresh = threshold(blurred.copy(),t)
            
plt.imshow(thresh,cmap='gray')
```

```python
def threshold(src_image,threshold):
    src_image[src_image>threshold] = 255
    src_image[src_image<255] = 0
    
    return src_image
```
