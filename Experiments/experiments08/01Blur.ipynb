```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline 

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import math
```

```python
image = cv2.imread("../../images/nature01.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image.shape
```

```python
plt.title("Original")
plt.imshow(image)
```

```python
blurred = np.hstack([
    cv2.blur(image, (3, 3))])

blurred.shape
```

```python
#fig = plt.figure(figsize=(5,5))
plt.title("Averaged")
plt.imshow(blurred)
```

### Average

```python
blurred = np.hstack([
    cv2.blur(image, (3, 3)),
    cv2.blur(image, (5, 5)),
    cv2.blur(image, (7, 7))])
```

```python
blurred.shape
```

```python
fig = plt.figure(figsize=(20,10))
plt.title("Averaged")
plt.imshow(blurred)
```

### GaussianBlur

```python
blurred = np.hstack([
    cv2.GaussianBlur(image, (3, 3), 0),
    cv2.GaussianBlur(image, (5, 5), 0),
    cv2.GaussianBlur(image, (7, 7), 0)])

fig = plt.figure(figsize=(15,10))
plt.title("GaussianBlur")
plt.imshow(blurred)
```

### Median

```python
blurred = np.hstack([
    cv2.medianBlur(image, 3),
    cv2.medianBlur(image, 5),
    cv2.medianBlur(image, 7)])

fig = plt.figure(figsize=(15,10))
plt.title("Median")
plt.imshow(blurred)
```

เทียบกับ average : median ได้ edge ที่ sharp กว่า

### Bilateral

https://en.wikipedia.org/wiki/Bilateral_filter

```python
blurred = np.hstack([
    cv2.bilateralFilter(image, 5, 21, 21),
    cv2.bilateralFilter(image, 7, 31, 31),
    cv2.bilateralFilter(image, 9, 41, 41)])

fig = plt.figure(figsize=(15,10))
plt.title("Bilateral")
plt.imshow(blurred)
```

```python
blurred = np.hstack([
    cv2.bilateralFilter(image, 9, 41, 41)])

fig = plt.figure(figsize=(8,8))
plt.title("Bilateral")
plt.imshow(blurred)
```
