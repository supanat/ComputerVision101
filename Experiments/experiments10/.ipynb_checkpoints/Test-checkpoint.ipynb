```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
image = cv2.imread("../../images/coins.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.title('Original')
plt.imshow(image,cmap='gray')
```

### Laplacian

```python
kernel = np.float32([[0, -1, 0],
                     [-1, 4, -1],
                    [0, -1, 0]])

dst = cv2.filter2D(image,-1,kernel)

plt.title('Laplacian Edge Detector')
plt.imshow(dst,cmap='gray')
```

```python
kernel = np.float32([[-1, -1, -1],
                     [-1, 8, -1],
                    [-1, -1, -1]])

dst = cv2.filter2D(image,-1,kernel)

plt.title('Laplacian Edge Detector Include Diagonals')
plt.imshow(dst,cmap='gray')
```

### Sobel

```python
# Horizontal
kernel = np.float32([[-1, 0, 1],
                     [-2, 0, 2],
                     [-1, 0, 1]])

sobel_horizontal = cv2.filter2D(image,-1,kernel)

plt.title('Sobel Edge Detector')
plt.imshow(sobel_horizontal,cmap='gray')
```

```python
# Vertical
kernel = np.float32([[1, 2, 1],
                     [0, 0, 0],
                     [-1, -2, -1]])

sobel_vertical = cv2.filter2D(image,-1,kernel)

plt.title('Sobel Edge Detector')
plt.imshow(sobel_vertical,cmap='gray')
```

```python
sobelCombined = cv2.bitwise_or(sobel_horizontal, sobel_vertical)
```

```python
plt.title('Sobel Edge Detector')
plt.imshow(sobelCombined,cmap='gray')
```
